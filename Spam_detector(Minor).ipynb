```python
#Ham spam filter
import pandas as pd
df=pd.read_csv("spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
#seeing the frequency of differnect categories
```

```python
#now convert the categorical data into numericals
#first convert the category col
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.Message,df.spam,test_size=0.25)
len(X_train)
```

```python
len(X_test)
```

We still have text in msg col this is changed using count vectorizer technique

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count=v.transform(X_test)
model.predict(X_test_count)
```

```python
model.score(X_test_count,y_test)
```

Using Sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
#Create pipeline using count vectorizer(convert text into some numericals)
#and apply naive classifier
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
